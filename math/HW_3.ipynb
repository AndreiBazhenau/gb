Нарисуйте график функции: y(x) = k∙cos(x – a) + b

для некоторых (2-3 различных) значений параметров k, a, b

```python
x = np.linspace(- 2*np.pi, 3*np.pi, 201)
plt.plot(x, 1 * np.cos(x - 0) + 0)
plt.plot(x, 2 * np.cos(x - 0) + 0, color='red')
plt.plot(x, 1 * np.cos(x - 2) + 0, color='green')
plt.plot(x, 1 * np.cos(x - 0) + 3, color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True);
```
