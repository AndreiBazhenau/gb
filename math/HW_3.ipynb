3. 1 Напишите код, который будет переводить полярные координаты в декартовы

```python
def pol_dec(angle, R):
    radian = angle * np.pi/180
    x = R * np.cos(radian)
    y = R * np.sin(radian)
    return x, y
    
# Проверка: angle = 90, R = 2
x, y = pol_dec(90, 2)
print(x, y)
```

4.1 Решите систему уравнений:

y = x**2 – 1

exp(x) + x*(1 – y) = 1

```python
def eq_root(z):
   x = z[0]
   y = z[1]
   F = np.empty((2))
   F[0] = x**2-y-1
   F[1] = exp(x)+x*(1-y)-1
   return F

z_pred = np.array([1,1])
z = fsolve(eq_root,z_pred)
print('x =', z[0],' ', 'y =', z[1])
```

4.2 Решите систему уравнений и неравенств:

y = x**2 – 1

exp(x) + x*(1 – y) > 1

```python
def eq_root(z):
   x = z[0]
   y = z[1]
   F = np.empty((2))
   F[0] = x**2-y-1
   F[1] = exp(x)+x*(1-y)-1.1
   return F

z_pred = np.array([1,1])
z = fsolve(eq_root,z_pred)
print('x =', z[0],' ', 'y =', z[1])
```
