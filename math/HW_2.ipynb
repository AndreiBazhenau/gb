2)	Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
def v_len(vector):
    return np.sqrt(vector[0]**2 + vector[1]**2 + vector[2]**2)
A = (10, 10, 10)
B = (0, 0, -10)
print(v_len(A))
print(v_len(B))
```

Почему прямые не кажутся перпендикулярными? (см.ролик)

Прфмые перпендикулярные если произведение их угловых коэффициентов = -1.

В данном примере условие выполняется, но у графика разный масштаб оси Х и оси Y.

Это можно исправить сделав график квадратным и установив лимиты x и y

```python
x = np.linspace(-5, 5, 21)
y = 3*x + 1
y2 = (-1/3)*x + 1
plt.figure(figsize=(4, 4))
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-15, 15)
plt.ylim(-15, 15);
```

Напишите код на Python, реализующий построение графиков:
1.	окружности,
2.	эллипса,
3.	гиперболы.

```python
x = []
y1 = []
y2 = []
x0 = 0
y0 = 0
R = 10
for i in range((-1000 * R), (1000 * R + 1)):
    x_ = i/1000
    x.append(x_)
    y1.append(math.sqrt(R**2 - (x_ - x0)**2) + y0)
    y2.append(-math.sqrt(R**2 - (x_ - x0)**2) + y0)
plt.figure(figsize=(4, 4))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim(-15, 15)
plt.ylim(-15, 15) 
plt.xlabel('x')
plt.ylabel('y');
```

```python
x = []
y1 = []
y2 = []
a = 10
b = 5
for i in range((-1000 * a), (1000 * a + 1)):
    x_ = i/1000
    x.append(x_)
    y1.append(b * math.sqrt(1 - x_**2/a**2))
    y2.append(-b * math.sqrt(1 - x_**2/a**2))
plt.figure(figsize=(4, 4))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim(-12, 12)
plt.ylim(-10, 10) 
plt.xlabel('x')
plt.ylabel('y');
```
