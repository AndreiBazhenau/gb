### 1. Общие

ALS
- На вход принимает только *позитивные* взаимодействия
- Прогнозирует **не** вероятность, а некоторые числа
- В ALS нельзя добавить внешние фичи
- ALS не учитывает сезонность / зависимость во времени (но это можно частично поправить - см далее)
- Можно использовать для ранжирования (как - см далее + вебинар 5)
- ALS учитывает только линейные взаимодействия между юзарами и товарами. Поэтому ALS в некотором смысле линейный алгоритм.

### 2. Что на самом деле оптимизирует ALS

$x_u^T$ - user embeddings  
$y_i$ - item embeddings  
$p_{ui}$ = 0 или 1. 1 - если в матрице user-item эдемент > 0 (было взаимодействие)  
$c_{ui}$ - Вес ошибки = элемент матрицы user-item  
$\lambda_x$, $\lambda_y$ - коэффициенты регуляризации  

**Что делает ALS**
- Максимизирует скалярное произведение (dot product) между векторами user-item для позитивных interactions, минимизирует для негативных

**Скалярное призведение (dot product) VS косинусная схожесть (cosine similarity)**

$$dot(a, b) = a * b = a_1*b_1 + a_2*b_2 + ...$$  
  
$$cos(a, b) = \frac{a * b}{||a||*||b||} = \frac{dot(a, b)}{||a||*||b||}$$

Косинусная схожесть имеет диапазон от -1 до 1 т.к. она отнормирована.

Похожесть эмбедингов измеряем с помощью косинусного расстояния.

### 3. Как учитывать изменение вкусов потребителей во времени / изменение популярности товара во времени

- Летом юзеры любят мороженое, а зимой - нет
- Многие фильмы популярны во время проката в кино, затем их популярность снижается

*Что делать?*  
- **Ограничить давность взаимодейстий**: обучаться не на всез 3-х годах датасета, а на последних нескольких месяцах. Недостаток - выбрасываем часть данных и теряем информацию. Поэтому используем следующий подход.
- **Давать недавним взаимодействиям больший вес**:  

$$value_w = \sum_i value_i * {\alpha^t}, 0 < \alpha < 1 $$

Это увеличивает метрику на 1-2%.

```python
import numpy as np

# Пример для 1 конкретного товара
#Если много покупок товара в начале - вес измениться сильно. Если в конце - слабо

# лист покупок по временной шкале: [сегодня, вчера, ... , год назад, ...]
monthly_purchase_list = [1, 0, 0, 3, 0, 1, 0, 0, 0, 2, 0, 0, 100, 0, 0, 3, 0, 1, 0, 0, 0, 2, 0, 10]  

def weight(input_list, alpha=0.9):
    """Чем меньше alpha, тем быстрее забывается история покупок"""
    
    assert (alpha > 0) & (alpha < 1), 'alpha should be in (0, 1)'
    
    for i, item in enumerate(input_list):
        input_list[i] = input_list[i] * (alpha**i) # i - кол-во месяцев, которое прошло с момента покупки товара
        
    return np.sum(input_list), input_list

# alpha - убывание веса прошлых покупок. как правило 0,7-0,8
weight(monthly_purchase_list, alpha=0.7)
```

### 4. Как использовать для ранжирования (Краткое введение. Подробности - на следующем вебинаре)

Ранжировать важно когда мы показываем товары в карусели по 3,5,10 товаров. В этом случае очень важно какие товары попадут в начало, потому что пользователь не видит все товары, а только первую часть.

**Вариант 1.**  
*As is* - ранжировать по output ALS

**Вариант 2.**
*Та же модель, другой лосс - BPR*

BPR - Bayesian Personalized Ranking loss
1. Случайная пара user - купленный им item_i = позитивная пара
2. Для того же юзера сэмплируем item_j (не купил или вес в user-item матрице ниже, у item_i из 1.) - негативная пара
3. Прогнозы модели $p_{ui}$, $p_{uj}$ --> $p_{ui} - p_{uj}$
4. loss = $\sigma(p_{ui} - p_{uj})$ - это и есть наша ошибка, нормированная через сигмоиду. Чем она больше, тем лучше.
5. Обновляем эмбеддинги

ALS пытается предсказать нолики и единицы. Если у пользователя много единичек (например 100) то ALSу трудно выбрать топ5 из 100.

BPR - это то же самое что и tf-idf, но с немного другими формулами для взвешивания.

```python
data = pd.read_csv('../../data/rec_sys/retail_train.csv')

data.columns = [col.lower() for col in data.columns]
data.rename(columns={'household_key': 'user_id',
                    'product_id': 'item_id'},
           inplace=True)


test_size_weeks = 3

data_train = data[data['week_no'] < data['week_no'].max() - test_size_weeks]
data_test = data[data['week_no'] >= data['week_no'].max() - test_size_weeks]

data_train.head(2)
```

```python
item_features = pd.read_csv('../../data/rec_sys/product.csv')
item_features.columns = [col.lower() for col in item_features.columns]
item_features.rename(columns={'product_id': 'item_id'}, inplace=True)

item_features.head(2)
```

```python
result = data_test.groupby('user_id')['item_id'].unique().reset_index()
result.columns=['user_id', 'actual']
result.head(2)
```

```python
n_items_before = data_train['item_id'].nunique()

data_train = prefilter_items(data_train) # не берёт take_n_popular=5000

n_items_after = data_train['item_id'].nunique()
print('Decreased # items from {} to {}'.format(n_items_before, n_items_after))
```

```python
user_item_matrix = pd.pivot_table(data_train, 
                                  index='user_id', columns='item_id', 
                                  values='quantity', # Можно пробовать другие варианты
                                  aggfunc='count', 
                                  fill_value=0
                                 )

user_item_matrix = user_item_matrix.astype(float) # необходимый тип матрицы для implicit

user_item_matrix.head(3)
```

Формируем словари из id матрицы в id товара и наоборот

```python
userids = user_item_matrix.index.values
itemids = user_item_matrix.columns.values

matrix_userids = np.arange(len(userids))
matrix_itemids = np.arange(len(itemids))

id_to_itemid = dict(zip(matrix_itemids, itemids))
id_to_userid = dict(zip(matrix_userids, userids))

itemid_to_id = dict(zip(itemids, matrix_itemids))
userid_to_id = dict(zip(userids, matrix_userids))
```

```python
user_item_matrix = bm25_weight(user_item_matrix.T).T  # Применяется к item-user матрице ! 
```

Сначала обучим ALS, а потом BayesianPersonalizedRanking и сравним

```python
%%time

model = AlternatingLeastSquares(factors=20, 
                                regularization=0.001,
                                iterations=15, 
                                calculate_training_loss=True, 
                                num_threads=4)

model.fit(csr_matrix(user_item_matrix).T.tocsr(),  # На вход item-user matrix
          show_progress=True)
```

```python
%%time

model_bpr = BayesianPersonalizedRanking(factors=20, 
                                regularization=0.001,
                                iterations=15,  
                                num_threads=4)

model_bpr.fit(csr_matrix(user_item_matrix).T.tocsr(),  # На вход item-user matrix
          show_progress=True)
```

```python
def get_recommendations(user, model, N=5):
    res = [id_to_itemid[rec[0]] for rec in 
                    model.recommend(userid=userid_to_id[user], 
                                    user_items=csr_matrix(user_item_matrix).tocsr(),   # на вход user-item matrix
                                    N=N, 
                                    filter_already_liked_items=False, 
                                    filter_items=[itemid_to_id[999999]],  # !!! 
                                    recalculate_user=False)]
    return res
```

```python
%%time
    
result['bm25_als'] = result['user_id'].apply(lambda x: get_recommendations(x, model=model, N=5))
result['bm25_bpr'] = result['user_id'].apply(lambda x: get_recommendations(x, model=model_bpr, N=5))
```

```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import cm
import pickle

from scipy.spatial.distance import cdist

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import seaborn as sns


def reduce_dims(df, dims=2, method='pca'):
    
    assert method in ['pca', 'tsne'], 'Неверно указан метод'
    
    if method=='pca':
        pca = PCA(n_components=dims)
        components = pca.fit_transform(df)
    elif method == 'tsne':
        tsne = TSNE(n_components=dims, learning_rate=250, random_state=42, n_iter=300, n_iter_without_progress=20)
        components = tsne.fit_transform(df)
    else:
        print('Error')
        
    colnames = ['component_' + str(i) for i in range(1, dims+1)]
    return pd.DataFrame(data = components, columns = colnames) 


def display_components_in_2D_space(components_df, labels='category', marker='D'):
    
    groups = components_df.groupby(labels)

    # Plot
    fig, ax = plt.subplots(figsize=(12,8))
    ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
    for name, group in groups:
        ax.plot(group.component_1, group.component_2, 
                marker='o', ms=6,
                linestyle='',
                alpha=0.7,
                label=name)
    ax.legend(loc='center left', bbox_to_anchor=(1.02, 0.5))

    plt.xlabel('component_1')
    plt.ylabel('component_2') 
    plt.show()
```
