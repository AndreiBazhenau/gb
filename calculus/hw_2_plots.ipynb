Нахождение предела функции ln(x)/x

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline

x = np.linspace(0, 1000, 1000)
plt.plot(x, np.log(x)/x)
plt.grid(True)
ylim(-1, 1)
xlim(0,100)
```

Исследование функции f(x) = x<sup>3</sup> + x<sup>2</sup>

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline

x = np.linspace(-10, 10, 1000)
y = x**3 - x**2
plt.plot(x, y)
plt.grid(True)
ylim(-1, 1)
xlim(-5,5)
```

1. Предложить пример функции, не имеющей предела в нуле и в бесконечностях.
Примером такой функции может быть y = sin(x+1/x). При x, стремящемся к 0 и бесконечностям, пределов не существует в силу периодичности синусоиды.

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline

x = np.linspace(-3, 3, 1000)
y = sin(x - 1/x)
plt.plot(x, y)
plt.grid(True)
#ylim(-50, 50)
#xlim(-10,10)
```

2. Пример функции, не имеющей предела в точке, но определённого в ней.

Можно взять сигнум-функцию у=sgn(x) в 0. Эта функция определена в 0, но не имеет в этой точке предела.

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline

x = np.linspace(-10, 10, 1000)
y = np.sign(x)
plt.plot(x, y)
plt.grid(True)
# ylim(-1, 1)
# xlim(-2,2)
```

Пример функции, не имеющей предела в бесконечностях

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline

x = np.linspace(-20, 20, 1000)
y = np.sin(2*x)/3*x
plt.plot(x, y)
plt.grid(True)
# ylim(-1, 1)
# xlim(-2,2)
```

Предложить пример функции, не имеющей предела в нуле и в бесконечностях.

y = (x<sup>2</sup> - 1)/x

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline

x = np.linspace(-2, 2, 1000)
y = (x**2 - 1)/x
plt.plot(x, y)
plt.grid(True)
ylim(-50, 50)
# xlim(-2,2)
```
