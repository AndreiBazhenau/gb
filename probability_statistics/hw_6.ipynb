ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].

Найдите ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy

Полученные значения должны быть равны.

Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков, а затем с использованием функций из библиотек numpy и pandas.

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
print(zp)
print(ks)
```

Ковариация двух величин, вычисленная с помощью элементраных действий:

```python
cov = np.mean(zp * ks) - np.mean(zp) * np.mean(ks)
cov
```

Ковариация с помощью функции cov из numpy

```python
np.cov(zp, ks, ddof=0)
```

Коэффициент корреляции Пирсона рассчитываем по формуле:

$$r_{XY} = \frac{cov_{XY}}{\sigma_{X}\sigma_{Y}}$$

Для этого находим стандратное отклонение sigma двух генеральных совокупностей:

```python
sigma_zp =  (np.sqrt(((zp - zp.sum() / len(zp))**2).sum() / len(zp)))
sigma_ks =  (np.sqrt(((ks - ks.sum() / len(ks))**2).sum() / len(ks)))
print(sigma_zp)
print(sigma_ks)
```

Проверяем правильность вычислений

```python
sigma_zp_= zp.std(ddof=0)
sigma_ks_ = ks.std(ddof=0)
print(sigma_zp_)
print(sigma_ks_)
```

Подставляем полученные значения в формулу и вычисляем коэффициент корреляции Пирсона:

```python
r = cov / (sigma_zp * sigma_ks)
r
```

Найдём коэффициент корреляции Пирсона с помощью функции NumPy и Pandas.

```python
r = np.corrcoef(zp, ks)
r
```

Найдите доверительный интервал для математического ожидания с надежностью 0.95.

Рассчитаем предельное отклонение от средней арифметической по следующей формуле:

$$T_{1,2} = \overline{X} \pm \frac{s_0}{\sqrt{n}} \cdot c_{\gamma}$$

где $T_1$, $T_2$ — нижняя и верхняя границы доверительного интервала,

$\overline{X}$  — выборочное среднее арифметическое, $s_0$ — среднее квадратичное отклонение по выборке (несмещенное),

$n$ — размер выборки,

$\gamma$ — доверительная вероятность,

$c_{\gamma} = Ф^{-1} \frac{(1 + \gamma)}{2}$ — обратное значение функции стандартного нормального распределения.

Итоговая формула: $$\left(\overline{x} - Z_{\frac{a}{2}} \cdot \frac{\sigma}{\sqrt{n}} , \overline{x} + Z_{\frac{a}{2}} \cdot \frac{\sigma}{\sqrt{n}}\right)$$

```python
iq = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])
n = len(iq)
t = 2.262
iq_mean = iq.sum() / len(iq)
print(iq_mean)
sigma =  (np.sqrt(((iq - iq.sum() / n)**2).sum() / (n - 1)))
print(sigma)
T1 = iq_mean - t * sigma / (n ** 0.5)
T2 = iq_mean + t * sigma / (n ** 0.5)
print(f'Доверительный интервал имеет границы ({T1}, {T2})')
```
